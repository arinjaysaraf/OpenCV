```python
# Cropping a Image
```

```python
img_bgr = cv2.imread("Mountain.jpeg",cv2.IMREAD_COLOR)
img_rgb = img_bgr[:,:,::-1]

plt.imshow(img_rgb)
print(img_rgb.shape)
```

```python
cropped_img = img_rgb[200:400, 300:600]
plt.imshow(cropped_img)
```

```python
desired_width = 100
aspect_ratio = desired_width / cropped_img.shape[1]
desired_height = int(cropped_img.shape[0] * aspect_ratio)
dim = (desired_width,desired_height)

resized_cropped = cv2.resize(cropped_img, dsize=dim, interpolation = cv2.INTER_AREA)
plt.imshow(resized_cropped)
```

```python
plt.imshow(img_rgb) 
```

```python
img_flipped_horz = cv2.flip(img_rgb,1)
img_flipped_vert = cv2.flip(img_rgb,0)
img_flipped_both = cv2.flip(img_rgb,-1)

plt.figure(figsize=[15,4])
plt.subplot(141);plt.imshow(img_flipped_horz);plt.title('Horizontal');
plt.subplot(142);plt.imshow(img_flipped_vert);plt.title('Vertical');
plt.subplot(143);plt.imshow(img_flipped_both);plt.title('Both');
plt.subplot(144);plt.imshow(img_rgb);plt.title('Original')
```

```python
imageLine = img_bgr.copy()

cv2.line(imageLine, (200,100),(400,100), (0,255,255),thickness = 5, lineType=cv2.LINE_AA);
cv2.line(imageLine, (200,200),(400,200), (0,255,255),thickness = 5, lineType=cv2.LINE_AA);
cv2.line(imageLine, (200,200),(200,100), (0,255,255),thickness = 5, lineType=cv2.LINE_AA);
cv2.line(imageLine, (400,200),(400,100), (0,255,255),thickness = 5, lineType=cv2.LINE_AA);


plt.imshow(imageLine[:,:,::-1])
```

```python
imageCircle = img_bgr.copy()

cv2.circle(imageCircle, (500,500), 100, (0,0,255),thickness = 5, lineType=cv2.LINE_AA);

plt.imshow(imageCircle[:,:,::-1]);
```

```python
imageRectangle = img_bgr.copy()

cv2.rectangle(imageRectangle, (300,300), (200,200), (255,0,255),thickness = 3 , lineType=cv2.LINE_AA);

plt.imshow(imageRectangle[:,:,::-1]);
```

```python
imageText = img_bgr.copy()
text = "This is the Himalayas"
fontScale = 2.2
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0,255,0)
fontThickness = 2

cv2.putText(imageText,text,(200,500),fontFace,fontScale,fontColor,fontThickness, cv2.LINE_AA);
plt.imshow(imageText[:,:,::-1])
```
